```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("NCRB-2021_Table.xlsx")
data = data.iloc[:, 1:]
data = data.set_index("Constituency name")
df = pd.DataFrame(data)

  
print(data)
```

```python
plt.plot(data.index,data['Murder (Sec.302 IPC) '])
plt.xlabel('State ')
plt.ylabel('Murder (Sec.302 IPC)')
plt.show()
```

```python
plt.plot(data.index,data['Dowry Deaths (Sec.304B IPC) '])
plt.xlabel('State ')
plt.ylabel('Dowry Deaths (Sec.304B IPC)')
plt.show()
```

```python
plt.plot(data.index,data['Robbery (Secs.392 to 394 IPC)'])
plt.xlabel('State ')
plt.ylabel('Robbery (Secs.392 to 394 IPC)')
plt.show()
```

```python
plt.plot(data.index,data['Rioting (Secs.147-151 & 153A IPC) '])
plt.rcParams['figure.figsize'] = [20,6]
plt.xlabel('State ')
plt.ylabel('Rioting (Secs.147-151 & 153A IPC) ')
plt.show()
```
